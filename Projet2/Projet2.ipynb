```python
import pandas as pd  # Manipulation des données
import numpy as np   # Calculs mathématiques
import matplotlib.pyplot as plt  # Visualisation
import seaborn as sns  # Visualisation avancée

# Afficher les graphiques directement dans le notebook
%matplotlib inline
```

```python
df = pd.read_csv("HRDataset_v14.csv")
df_copy = df.copy()
df.head()
```

```python
df.info()
```

```python
df[['month', 'day', 'year']] = df['DOB'].str.split('/', expand=True)
df['year'] = df['year'].astype(int) + 1900
df['DOB'] = pd.to_datetime(df[['month', 'day', 'year']])
```

```python
df["DateofTermination"] = pd.to_datetime(df["DateofTermination"])

df["LastPerformanceReview_Date"] = pd.to_datetime(df["LastPerformanceReview_Date"])

df['DateofHire'] = pd.to_datetime(df['DateofHire'])

df["AgeHired"] = df["DateofHire"].dt.year - df["DOB"].dt.year

today = pd.to_datetime('today')
df["CurrentAge"] = today.year - df["DOB"].dt.year

def calculate_tenure(row):
    if pd.notna(row['DateofTermination']):
        return row['DateofTermination'].year - row['DateofHire'].year
    else:
        return row['CurrentAge'] - row['AgeHired']

# Appliquer la fonction à chaque ligne
df['TenureYears'] = df.apply(calculate_tenure, axis=1)
```

```python
sorted_column = ["Employee_Name", "EmpID", "DOB", "CurrentAge", "Sex", "GenderID", "MarriedID", "MaritalStatusID", "MaritalDesc", "CitizenDesc", "HispanicLatino", "RaceDesc", "Position", "PositionID", "Department", "DeptID", "EmpStatusID", "EmploymentStatus", "DateofHire", "AgeHired", "TenureYears", "Salary", "PerfScoreID", "PerformanceScore", "LastPerformanceReview_Date", "EngagementSurvey", "EmpSatisfaction", "SpecialProjectsCount", "DaysLateLast30", "Absences", "ManagerName", "ManagerID", "RecruitmentSource", "FromDiversityJobFairID", "Termd", "DateofTermination", "TermReason", "State", "Zip"]
df_sorted = df[sorted_column]
df_sorted.head(10)
```

```python
numerical_analysis_vars  = ["CurrentAge", "AgeHired", "TenureYears", "Salary", "PerfScoreID", "PerformanceScore", "EngagementSurvey", "EmpSatisfaction", "SpecialProjectsCount", "DaysLateLast30", "Absences"]
df_analysis = df[numerical_analysis_vars]
df_analysis.describe().round(2)
```

```python
mean_tenure = df_analysis["TenureYears"].mean()
std_tenure = df_analysis["TenureYears"].std()

def assign_group(row):
    if row["TenureYears"] < mean_tenure - std_tenure:
        return 0  # Sous la moyenne - écart-type
    elif row["TenureYears"] < mean_tenure:
        return 1  # Moyenne - écart-type à moyenne
    elif row["TenureYears"] < mean_tenure + std_tenure:
        return 2  # Moyenne à moyenne + écart-type
    else:
        return 3  # Au-dessus de la moyenne + écart-type

# Appliquer la fonction à chaque ligne de la dataframe df_analysis
df_analysis['TenureGroup'] = df_analysis.apply(assign_group, axis=1)

# Afficher les résultats
print(df_analysis['TenureGroup'])
```
